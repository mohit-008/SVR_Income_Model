```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('income.csv') 
df.head()
```

```python
df.info()
```

```python
for i in df.columns:
    if df[i].dtype != 'int64':
        print(i,':',df[i].unique())
```

```python
df.columns
```

```python
data = pd.get_dummies(df,columns=['Education_Level', 'Occupation',
       'Location', 'Marital_Status', 'Employment_Status',
       'Homeownership_Status', 'Type_of_Housing', 'Gender',
       'Primary_Mode_of_Transportation'],dtype='int',drop_first=True)
```

```python
data.head()
```

```python
# sns.pairplot(data)
# plt.show()
```

```python
sns.scatterplot(data=data,x="Education_Level_Master's",y='Income')
plt.show()
```

```python
data.corr()
```

```python
X = data.drop('Income',axis=1)
y = data['Income']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
sc.fit(X)
```

```python
X_trans = sc.transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,y,test_size=0.2,random_state=101)
```

```python
from sklearn.svm import SVR
model = SVR(kernel='rbf')
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
print('R2 score',r2_score(Y_test,y_pred)*100)
print('RMSE',root_mean_squared_error(Y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
L = LinearRegression()
L.fit(X_train,Y_train)
y_pred_linear = L.predict(X_test)
```

```python
print('R2 score',r2_score(Y_test,y_pred_linear)*100)
print('RMSE',root_mean_squared_error(Y_test,y_pred_linear)) 
```

```python
from sklearn.tree import DecisionTreeRegressor
DTR = DecisionTreeRegressor(max_depth=6)
DTR.fit(X_train,Y_train)
y_pred_tree = DTR.predict(X_test)
```

```python
print('R2 score',r2_score(Y_test,y_pred_tree)*100)
print('RMSE',root_mean_squared_error(Y_test,y_pred_tree)) 
```
